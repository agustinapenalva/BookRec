```python
import kagglehub
import pandas as pd
import os

#DOWNLOAD "BOOK RECOMMENDATION DATASET" FROM KAGGLE

# Download latest version of the dataset
path = kagglehub.dataset_download("arashnic/book-recommendation-dataset")

# Print the path to the dataset files
print("Path to dataset files:", path)

# List all files in the dataset directory
dataset_files = os.listdir(path)

# Load each file into a DataFrame and store them in a dictionary
dfs = {}

for file in dataset_files:
    file_path = os.path.join(path, file)

    # Check the file type (if it's a CSV, we can load it into a DataFrame)
    if file.endswith('.csv'):
        dfs[file] = pd.read_csv(file_path)
        print(f"Loaded DataFrame for {file}")

# Now `dfs` contains DataFrames for each CSV file
```
